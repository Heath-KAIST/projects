# KSIF Library

```python
from ksif import Portfolio
from ksif.core.columns import *
```

```python
pf = Portfolio(start_date='2008-01-01')
pf
```

__benckmark 설정__

```python
print(pf.benchmark)
pf.set_benchmark('코스닥')
pf.benchmark
```

__포트폴리오 연산__

```python
pf.outcome(weighted=PBR, show_plot=True)
```

```python
ks = pf.loc[pf[EXCHANGE] == '유가증권시장', :]  # KOSPI 종목만 선택
ks.periodic_rank(min_rank=1, max_rank=30, factor=ROE)    # ROE 기준으로 1등부터 30등까지 매달 투자
# ks.periodic_rank(min_rank=1, max_rank=30, factor=ROE, bottom=True).outcome(show_plot=True)
```

```python
kq = pf.loc[pf[EXCHANGE] == '코스닥', :]    # KOSDAQ 종목만 선택
kq.periodic_percentage(min_percentage=0.1, max_percentage=0.2, bottom=True).outcome(show_plot=True)    # MKTCAP 내림차순으로 정렬해 10~20% 매달 투자
```

```python
# 상장된 전체 기업을 6개월 모멘텀을 기준으로 10등분해 동일비중으로 10분위 누적수익률
pf.quantile_distribution_ratio(factor=MOM6, chunk_num=10, show_plot=True, show_bar_chart=True)
```
